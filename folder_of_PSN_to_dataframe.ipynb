```python
#Descomprimir archivos formato .gz
import os
import gzip
import shutil
import numpy as np
import pandas as pd
#import seaborn as sns
#sns.set()
from matplotlib import pyplot as plt
import datetime
import time            
            
#devolverArchivos("/home/gfajardo/Desktop/02-Feb/")

# import module sys to get the type of exception

import sys


#descomprimirArchivos("/home/gfajardo/Desktop/Datos 121001/archive/2020/")

def devolverArchivos(carpeta):
    
    count=0
    tabla2=[]
    for archivo2 in os.listdir(carpeta):
        
            os.chdir(carpeta+archivo2)
            print(os.path.join(carpeta,archivo2))
            for archivo in os.listdir(carpeta+archivo2):
                try:
                    if count<1:
                        print(os.path.join(carpeta,archivo2,archivo))
                        tabla=posiciontiempo(os.path.join(carpeta,archivo2,archivo))


                    else:
                        print(os.path.join(carpeta,archivo2,archivo))
                        tabla=posiciontiempo(os.path.join(carpeta,archivo2,archivo))
                        tabla=pd.concat([tabla,tabla2], axis=0)
                        tabla=tabla.reset_index(drop=True)

                    #pd.set_option('display.max_rows', 8000)       
                    tabla2=tabla
                    count=count+1
                except:
                        print("Oops!", sys.exc_info()[0], "occurred.")
                        print("Next entry.")
                        print()
    return(tabla2)
            
          

            
Mli=devolverArchivos("/home/gfajardo/Desktop/Datos 121001/archive/2018/")
Mli.to_csv(r'/home/gfajardo/Desktop/Datos 121001/PSN2018conAÃ‘O.csv')
```

```python
import os
import gzip
import shutil
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
from matplotlib import pyplot as plt
import datetime
import time
data_file='/home/gfajardo/Desktop/151231_000000.psn'
#column_names=['AÃ±o','Day','Tiempo','PRN','S4','Az','Elv']



def posiciontiempo(data_file):
    
    Mli = pd.read_csv(data_file, header=None)#names=column_names)#)
    #Mli = pd.read_csv(data_file, header=None, delim_whitespace=True)
    #Mli = Mli.drop(Mli.columns[[0]], axis=1)
 
    # Delimiter
    data_file_delimiter = '\s+'
        
    # The max column count a line in the file could have
    largest_column_count = 0
    
    # Loop the data lines
    with open(data_file, 'r') as temp_f:
            # Read the lines
            lines = temp_f.readlines()
    
            for l in lines:
                # Count the column count for the current line
                column_count = len(l.split())+ 1
    
                # Set the new most column count
                largest_column_count = column_count if largest_column_count < column_count else largest_column_count
    
    # Close file
    temp_f.close()
    
     # Generate column names (will be 0, 1, 2, ..., largest_column_count - 1)
    column_names = [i for i in range(3, largest_column_count)]
    #print(column_names)
    column_names = ['Lat','Long','Alt']+column_names
    
    print(column_names)

    # Read csv
    Mli = pd.read_csv(data_file, header=None, delim_whitespace=True, names=column_names)
    #print(range(2,5))
    #print(len(Mli))
    #print(type(Mli))
    Time=Mli[Mli['Lat'] == "T"]
    #print(Time.drop(Time[2],axis=1))
    Time = Time.drop(Time.columns[range(5,largest_column_count)], axis=1)  # df.columns is zero-based pd.Index 
    Time['index_col'] = Time.index
    #print('Tabla Time',Time)
    onlypositions= Mli[Mli['Lat'] != "T"]
    onlypositions = onlypositions.drop(onlypositions.columns[range(3,largest_column_count)], axis=1)  # df.columns is zero-based pd.Index 
    #print(onlypositions)
    
    onlypositions  = onlypositions .reset_index(drop=True)
    #print(Mli)
    #print(onlypositions)
    
    #print(len(Mli))
    
    #Version 3    
    a=[] 
    y=0
    dia=[]
    w=[]
    df_marks = pd.DataFrame(columns=['AÃ±o','Dia','Seg']) 
    for i in Time.index:
        x=i+1
        z=x-y-1
        for j in range (0,z-1):
            
            n=int(Time[4][i])-((z-1)-j)
            #print((z-1)-j)
            m=int(Time[3][i])
            aÃ±o=Time['Long'][i]
            
            
            DOY=datetime.date(int(Time['Long'][i])+2000, int(Time['Alt'][i]), int(Time[3][i]))
            mD=int(DOY.strftime('%j'))
            
            if n<0: 
                dia+=[m-1]
                a+=[n+86400]
    
                new_row = { 'AÃ±o':aÃ±o,'Dia':mD-1,'Seg':n+86400}
                #append row to the dataframe
                df_marks = df_marks.append(new_row, ignore_index=True)
            else:
                dia+=[m]
                a+=[n]
                w+=[m,n]
                new_row = { 'AÃ±o':aÃ±o,'Dia':mD,'Seg':n}
                #append row to the dataframe
                df_marks = df_marks.append(new_row, ignore_index=True)
            #print(j)
        y=i 


    #Con este cÃ³digo se asegura completar la tabla si existen datos despuÃ©s de una marca temporal:

    it= range(0,len(Mli)-y)
    #print(range(6,46))
    #print(len(Mli))
    #print('ultimo:')
    #print(it)
    aux=1
    for x in it:  
    
                n=int(Time[4][i])+aux 
                m=int(Time[3][i])
                DOY=datetime.date(int(Time['Long'][i])+2000, int(Time['Alt'][i]), int(Time[3][i]))
                mD=int(DOY.strftime('%j'))
                aÃ±o=Time['Long'][i]
                
                dia+=[m]
                a+=[n]
                w+=[m,n]
                new_row = { 'AÃ±o':aÃ±o,'Dia':mD,'Seg':n}
                #append row to the dataframe
                df_marks = df_marks.append(new_row, ignore_index=True)
                aux+=1
    #print(y)
    #pd.set_option('display.max_rows', 400)
    #print(df_marks)
    #print(len(df_marks))
    result = pd.concat([onlypositions, df_marks], axis=1)
    pd.set_option('display.max_rows', 350)
    
    return(result) 


tabla2=posiciontiempo('/home/gfajardo/Desktop/01-Jan/151231_010000.psn')
tabla2.to_csv(r'/home/gfajardo/Desktop/Datos 121001/PSN2015conAÃ‘O.csv.csv')
```
